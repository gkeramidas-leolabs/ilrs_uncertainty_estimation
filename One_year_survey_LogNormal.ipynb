```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
plt.style.use('ggplot')
plt.rcParams['font.size'] = 20
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['axes.labelweight'] = 'heavy'
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['figure.titlesize'] = 12
plt.rcParams['text.usetex']=True
```

```python
months = ["","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec","Jan","Feb","Mar","Apr","May"]
```

```python
mlist = [x+1 for x in range(13)]
```

```python
def z_score(d, MAD):
    return 0.6745*d/MAD

def cull_outliers(arr):
    threshold = 3.0

    arr = np.array(arr)

    # Strip nans
    arr = arr[~np.isnan(arr)]

    diffs = np.sqrt((arr - np.median(arr))**2)

    MAD = np.median(diffs) 
    
    new_arr = []
    
    for i in range(len(arr)):
        if z_score(diffs[i],MAD) < threshold:
            new_arr.append(arr[i])
        else:
            new_arr.append(np.median(arr))

    return new_arr
```

## AJISAI

```python
filepath_aj = "/Users/gkeramidas/Projects/learning/L3059-uncertainties-LN.txt"
```

```python
dx_aj_h = []
dy_aj_h = []
dz_aj_h = []
dx_aj_s = []
dy_aj_s = []
dz_aj_s = []
```

```python
with open(filepath_aj,"r") as rf:
    next(rf)
    for columns in (raw.strip().split() for raw in rf):
        dx_aj_h.append(float(columns[1]))
        dy_aj_h.append(float(columns[2]))
        dz_aj_h.append(float(columns[3]))
        dx_aj_s.append(float(columns[4]))
        dy_aj_s.append(float(columns[5]))
        dz_aj_s.append(float(columns[6]))
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,5))
ax1.plot(mlist,cull_outliers(dx_aj_h),"sb-",label="hts")
ax1.plot(mlist,cull_outliers(dx_aj_s),"sr-",label="sgf")
ax2.plot(mlist,cull_outliers(dy_aj_h),"sb-",label="hts")
ax2.plot(mlist,cull_outliers(dy_aj_s),"sr-",label="sgf")
ax3.plot(mlist,cull_outliers(dz_aj_h),"sb-",label="hts")
ax3.plot(mlist,cull_outliers(dz_aj_s),"sr-",label="sgf")

ax1.set_ylabel(r"$\delta X$ (m)")
ax2.set_ylabel(r"$\delta Y$ (m)")
ax3.set_ylabel(r"$\delta Z$ (m)")
ax1.xaxis.set_major_locator(MaxNLocator(14))
ax1.set_xticklabels(months)
ax2.xaxis.set_major_locator(MaxNLocator(14))
ax2.set_xticklabels(months)
ax3.xaxis.set_major_locator(MaxNLocator(14))
ax3.set_xticklabels(months)
ax1.set_title("AJISAI")
ax1.legend()
plt.show()
```
