```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
plt.style.use('ggplot')
plt.rcParams['font.size'] = 20
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['axes.labelweight'] = 'heavy'
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['figure.titlesize'] = 12
plt.rcParams['text.usetex']=True
```

```python
months = ["","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec","Jan","Feb","Mar","Apr","May"]
```

```python
mlist = [x+1 for x in range(13)]
```

Read the multiprovider data first

```python
filepath_lrt_mp = "/Users/gkeramidas/Projects/learning/L2486-uncertainties.txt"

dx_lrt_h_mp = []
dy_lrt_h_mp = []
dz_lrt_h_mp = []
with open(filepath_lrt_mp,"r") as rf:
    next(rf)
    for columns in (raw.strip().split() for raw in rf):
        dx_lrt_h_mp.append(float(columns[1]))
        dy_lrt_h_mp.append(float(columns[2]))
        dz_lrt_h_mp.append(float(columns[3]))
```

Read single provider data

```python
filepath_lrt_sp = "/Users/gkeramidas/Projects/learning/L2486-SP-uncertainties.txt"
```

```python
dx_lrt_h_1d = []
dy_lrt_h_1d = []
dz_lrt_h_1d = []
dx_lrt_h_lin = []
dy_lrt_h_lin = []
dz_lrt_h_lin = []
dx_lrt_h_exp = []
dy_lrt_h_exp = []
dz_lrt_h_exp = []

with open(filepath_lrt_sp,"r") as rf:
    next(rf)
    for columns in (raw.strip().split() for raw in rf):
        dx_lrt_h_1d.append(float(columns[1]))
        dy_lrt_h_1d.append(float(columns[2]))
        dz_lrt_h_1d.append(float(columns[3]))
        dx_lrt_h_lin.append(float(columns[4]))
        dy_lrt_h_lin.append(float(columns[5]))
        dz_lrt_h_lin.append(float(columns[6]))
        dx_lrt_h_exp.append(float(columns[7]))
        dy_lrt_h_exp.append(float(columns[8]))
        dz_lrt_h_exp.append(float(columns[9]))
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,5))
ax1.plot(mlist,dx_lrt_h_mp,"sb-",label="MP")
ax1.plot(mlist,dx_lrt_h_1d,"sr-",label="1d")
#ax1.plot(mlist,dx_lrt_h_lin,"sg-",label="lin")
ax1.plot(mlist,dx_lrt_h_exp,"sm-",label="exp")
ax2.plot(mlist,dy_lrt_h_mp,"sb-",label="MP")
ax2.plot(mlist,dy_lrt_h_1d,"sr-",label="1d")
#ax2.plot(mlist,dy_lrt_h_lin,"sg-",label="lin")
ax2.plot(mlist,dy_lrt_h_exp,"sm-",label="exp")
ax3.plot(mlist,dz_lrt_h_mp,"sb-",label="MP")
ax3.plot(mlist,dz_lrt_h_1d,"sr-",label="1d")
#ax3.plot(mlist,dz_lrt_h_lin,"sg-",label="lin")
ax3.plot(mlist,dz_lrt_h_exp,"sm-",label="exp")

ax1.set_ylabel(r"$\delta X$ (m)")
ax2.set_ylabel(r"$\delta Y$ (m)")
ax3.set_ylabel(r"$\delta Z$ (m)")
ax1.xaxis.set_major_locator(MaxNLocator(14))
ax1.set_xticklabels(months)
ax2.xaxis.set_major_locator(MaxNLocator(14))
ax2.set_xticklabels(months)
ax3.xaxis.set_major_locator(MaxNLocator(14))
ax3.set_xticklabels(months)
ax1.set_title("LARETS")
ax1.legend()
plt.show()
```
